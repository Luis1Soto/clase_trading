```python
import pandas as pd
import numpy as np

data  = pd.read_csv('aapl_1s_data.csv')
```

```python
print(data)
```

```python
data['Returns'] = np.log(data['Close'] / data['Close'].shift(1))
cov_log_returns = data['Returns'].cov(data['Returns'].shift(1))
spread_roll_log = 2 * np.sqrt(-cov_log_returns)
spread_roll_log
```

```python
data['Returns'] = data['Close'] / data['Close'].shift(1) -1
cov_returns = data['Returns'].cov(data['Returns'].shift(1))
spread_roll = 2 * np.sqrt(-cov_returns)
spread_roll
```

```python
spread_roll_values = []
data['Returns'] = data['Close'] / data['Close'].shift(1) -1
for i in range(1, len(data) - 1):
    cov_returns_iter = data['Returns'].iloc[i:i+2].cov(data['Returns'].shift(1).iloc[i:i+2])
    spread_iter = 2 * np.sqrt(-cov_returns_iter) if cov_returns_iter < 0 else 0
    spread_roll_values.append(spread_iter)

pd.Series(spread_roll_values)
```

```python
pd.Series(spread_roll_values).mean()
```
